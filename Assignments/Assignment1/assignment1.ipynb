### Task 1: A list of dictionaries (1p)



````python
#The list of raw transaction records given:
transactions = [
    {"id": 1, "product": "Laptop", "price": 1200, "quantity": 2, "region": "EU"},
    {"id": 2, "product": "Phone", "price": 800, "quantity": -1, "region": "US"},
    {"id": 3, "product": "Tablet", "price": 0, "quantity": 5, "region": "EU"},
    {"id": 4, "product": "Monitor", "price": 350, "quantity": 3, "region": "US"},
    {"id": 5, "product": "Laptop", "price": 1200, "quantity": 1, "region": "EU"},
    {"id": 6, "product": "Keyboard", "price": 75, "quantity": 10, "region": "APAC"},
]

#print(transactions)
#type(transactions)

#a) Filter out invalid transactions where the price is zero or the quantity is negative.

#I filter based on the rules and create a new list of dictionaries which contain only the ones 
#which fulfil the criteria.
transactions_new =[]   #create an empty list which is then filled next
for trans in transactions:
    if trans["price"] > 0 and trans["quantity"]>=0:
        transactions_new.append(trans)

print("filtered out invalid transactions:")
print(transactions_new)
    
````

````python
#b) For each remaining transaction, compute a new field `total` = `price * quantity`.
# I perform this creating a new field by looping across each dictionary in the list 
for trans_new in transactions_new:
    trans_new["total"]= trans_new["price"] * trans_new["quantity"]

for trans_new in transactions_new:
    print(trans_new)


````

````python
#c) Group the valid transactions by `region` and compute the total revenue per region. 
#Store the result in a dictionary.

regional_revenue = {} # this creates an empty dictionary
for reg in transactions_new: #this goes through each item in the transactions_new list
    if reg["region"] not in regional_revenue: #this checks whether the region exists in the new dictionary
        regional_revenue[reg["region"]]= 0   #
        regional_revenue[reg["region"]] +=reg["total"]

print(regional_revenue)
    
````

````python
# d) Print the region with the highest total revenue and its value using an f-string.  

def get_value(k): 
    return regional_revenue[k]
    
max_region = max(regional_revenue, key=get_value)
#print(max_region)

print(f"\nRegion with the highest total revenue is {max_region} with total value of {regional_revenue[max_region]}")
````

### Task 2: Creating a function (1p)



````python
# the list of daily sales figures (in euros) for a two-week period given:

sales = [1200, 980, 1540, 870, 2010, 1650, 1300, 1100, 950, 1780, 2200, 1430, 1600, 1900]

#print(sales)
#type(sales)

#Write a function called summarise_sales that takes a list of daily sales values and returns a dictionary containing:

from statistics import mean 

#I have used R a lot and was surprised to learn that Python does not a built-in mean
#function. I took the easy way out and imported it from the statistics module. 
#However, I also calculated average manually. len function is the same as in R.

def summarise_sales(x):
    total = sum(x)
    average = mean(x)
    average_manual= sum(x)/len(x)
    best_day = x.index(max(x))+1
    worst_day = x.index(min(x))+1
    above_average_count = len([day for day in x if day > average])
    return {"total": total, "average": average, "average manual": average_manual, "best_day": best_day, "worst_day": worst_day, "above_average_count": above_average_count}
  
result = summarise_sales(sales)

print(result)
type(result)
````

### Task 4: Class and methods (1p)



````python
#The constructor provided

class Portfolio:
    def __init__(self, owner):
        self.owner = owner
        self.holdings = {}  # ticker -> {shares, price_per_share}

#  a) A method add_asset(ticker, shares, price_per_share) that adds a new asset or updates the number 
    #of shares if the ticker already exists (in which case the price should also be updated).  
    
    def add_asset(self, ticker, shares, price_per_share):
        #self.holdings[ticker] = {"shares": shares,  "price_per_share": price_per_share} #creates a dictionary 
        # which is used below to create a dictionary always when the ticker is not in holdings: 
        if ticker not in self.holdings: #checks if the ticker already exists
            self.holdings[ticker] ={"shares": shares,  "price_per_share": price_per_share}
        else:
            self.holdings[ticker]["shares"] =shares
            self.holdings[ticker]["price_per_share"] =price_per_share               

#b) A method total_value() that returns the total market value of the portfolio.
    def total_value(self):
        total_value = 0
        for ticker in self.holdings:
            asset_value = self.holdings[ticker]["shares"] * self.holdings[ticker]["price_per_share"]
            total_value += asset_value
        return total_value


#c) A method most_valuable() that returns 
#the ticker of the asset with the highest total value (shares × price).
    def most_valuable(self):
        most_valuable_asset = max(self.holdings, key=lambda ticker: 
                                  self.holdings[ticker]["shares"] * self.holdings[ticker]["price_per_share"])
        return most_valuable_asset
        
      
#d) A __str__ method that returns a readable summary of the portfolio, 
#including the owner's name, the number of assets, and the total value.
    def __str__(self):
        return f"Owner: {self.owner}, Assets: {len(self.holdings)},  Total value: {self.total_value()}"



#Create an instance, add at least three assets, and demonstrate all methods with printed output.

#creating a new instance called portfolio_new where "test" is stored as self.owner. So now Python knows to use
#portfolio_new in this class. So self= portfolio_new

portfolio_new = Portfolio("test")

#Adding three assets with the add_asset method:
portfolio_new.add_asset("ABCD", 10, 100)
portfolio_new.add_asset("EFGH", 20, 140)
portfolio_new.add_asset("IJKL", 30, 160)

print("Total value of portfolio:")
print(portfolio_new.total_value())

print("")

print("Most valuable asset:")
print(portfolio_new.most_valuable())

print("")
#this is using the __str__ method to print a summary of the portfolio:

print(portfolio_new)   
````

### Task 8: More Pandas (1p)



````python
import numpy as np
import pandas as pd

rng = np.random.default_rng(seed=123)
days = pd.date_range(start="2025-01-06", periods=28, freq="D")
visitors = rng.integers(200, 600, size=28)
conversion_rate = rng.uniform(0.02, 0.08, size=28).round(4)
avg_order_value = rng.normal(50, 10, size=28).round(2)
````

````python
#a) Create a DataFrame with columns: `date`, `visitors`, `conversion_rate`, and `avg_order_value`. 
#Set `date` as the index.

data={"date": days, "visitors": visitors, "conversion_rate": conversion_rate, "avg_order_value": avg_order_value}
data_df=pd.DataFrame(data)
#data_df

#setting the date as index so instead of 0,1,2... we use the date
data_df = data_df.set_index("date")
data_df

#b) Compute two new columns: `conversions` (= visitors × conversion_rate, 
#rounded to the nearest integer) and `daily_revenue` (= conversions × avg_order_value).

data_df["conversions"]=(data_df["visitors"]*data_df["conversion_rate"]).round(0)
data_df["daily_revenue"]=data_df["conversions"]*data_df["avg_order_value"]

data_df
````
